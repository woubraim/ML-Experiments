# <font color='cadetblue'>1- Import the dataset</font>



```python
dataset = pd.read_csv(r"data_.csv")
```

```python
df = pd.DataFrame(dataset)
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

# <font color='dodgerblue'>2- Identifying and handling the missing values</font>


```python
df.isnull().sum()
```

# <font color='lime'>Dropna</font>

```python
df1 = df.copy()
```

```python
# summarize the shape of the raw data
print("Before:",df1.shape)

# drop rows with missing values
df1.dropna(inplace=True)

# summarize the shape of the data with missing rows removed
print("After:",df1.shape)
```

# <font color='chartreuse'>Fillna</font>

```python
df2 = df.copy()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# fill missing values with mean column values
df2.fillna(df2.mean(), inplace=True)
# count the number of NaN values in each column
print(df2.isnull().sum())

df2
```

# <font color='darkgreen'>Scikit-Learn</font>



```python
X
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
print(X)
```

# <font color='royalblue'>3- Encoding the categorical data</font>


# <font color='darkorchid'>ColumnTransformer</font>

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
df
```

```python
print(X)
```

# <font color='magenta'>LabelEncoder</font>

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

# <font color='aqua'>4- Splitting the dataset</font>


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

# <font color='lightskyblue'>4- Feature scaling</font>


```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
X_train[:, 3:] = mm.fit_transform(X_train[:, 3:])
X_test[:, 3:] = mm.transform(X_test[:, 3:])
```

```python
print(X_train[:, 3:])
```

```python
print(X_test[:, 3:])
```

# <font color='red'>Standard Scaler</font>

```python
from sklearn.preprocessing import StandardScaler
sta = StandardScaler()
X_train[:, 3:] = sta.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sta.transform(X_test[:, 3:])
```
